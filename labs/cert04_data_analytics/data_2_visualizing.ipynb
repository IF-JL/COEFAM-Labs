Uses a shared synthetic **factory production log** — one row per part made across 3 presses, 2 shifts, 5 days. pandas, NumPy and matplotlib are pre-installed in Colab, so there is nothing to install.

## Objectives
By the end you will be able to:
- Plot a process trend over time.
- Show a distribution with a histogram.
- Compare categories with a bar chart and a box plot.

## Load

```python
import pandas as pd
DATA = "https://raw.githubusercontent.com/IF-JL/COEFAM-Labs/main/labs/cert04_data_analytics/data/"
df = pd.read_csv(DATA + "production_log.csv", parse_dates=["timestamp"])
df.head()
```

## 1 · Line plot — a process trend
One machine, one shift: how did oven temperature move over time?

```python
import matplotlib.pyplot as plt
one = df[(df.machine=="Press_C") & (df.date=="2020-06-01") & (df.shift=="Day")]
plt.figure(figsize=(9,3))
plt.plot(one["timestamp"], one["oven_temp_c"])
plt.title("Press_C oven temperature — 2020-06-01 Day"); plt.ylabel("deg C"); plt.show()
```

## 2 · Histogram — a distribution

```python
plt.figure(figsize=(7,3))
plt.hist(df["cycle_time_s"], bins=40)
plt.title("Cycle-time distribution"); plt.xlabel("seconds"); plt.ylabel("count"); plt.show()
```

## 3 · Bar chart — compare categories

```python
fail_rate = df.groupby("machine")["result"].apply(lambda s: (s=="Fail").mean())
plt.figure(figsize=(5,3))
plt.bar(fail_rate.index, fail_rate.values)
plt.title("Fail rate by machine"); plt.ylabel("fail rate"); plt.show()
```

## 4 · Box plot — distribution by group
See spread and outliers per machine at a glance.

```python
names = ["Press_A", "Press_B", "Press_C"]
data = [df[df.machine==m]["oven_temp_c"] for m in names]
plt.figure(figsize=(6,3))
plt.boxplot(data)                 # version-robust: set labels via xticks
plt.xticks([1, 2, 3], names)
plt.title("Oven temperature by machine"); plt.ylabel("deg C"); plt.show()
```
