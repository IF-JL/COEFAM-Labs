## 1 · Load the data

```python
import pandas as pd
DATA = "https://raw.githubusercontent.com/IF-JL/COEFAM-Labs/main/labs/cert04_data_analytics/data/"
df = pd.read_csv(DATA + "production_log.csv", parse_dates=["timestamp"])
df.head()
```

## 2 · Inspect it
Shape, column types, and quick statistics.

```python
print("rows, columns:", df.shape)
df.info()
df.describe()
```

## 3 · Select columns & filter rows

```python
# a few columns
df[["machine", "shift", "oven_temp_c", "result"]].head()
```

```python
# rows: only Press_C, and all failures
press_c = df[df["machine"] == "Press_C"]
fails   = df[df["result"] == "Fail"]
print("Press_C parts:", len(press_c), "| total failures:", len(fails))
```

## 4 · Group & aggregate
The core analytics move: split by a category, compute a number.

```python
# parts made per machine
print(df.groupby("machine").size(), "\n")

# fail rate per machine
fail_rate = df.groupby("machine")["result"].apply(lambda s: (s == "Fail").mean())
print("Fail rate by machine:")
print(fail_rate.round(3))
```
